```python
import os
import shutil
import pandas as pd
```

```python
# разархивирование датасета со стилями
!unzip /content/drive/MyDrive/SF/Project_4_Style_Transfer/wikiart.zip
!unzip /content/drive/MyDrive/SF/Project_4_Style_Transfer/wikiart_csv.zip
```

```python
# сохранение путей к изображениям для валидации в paths_valid
df_style=pd.read_csv('style_val.csv', header=None, names = ['path', 'style_index'])
paths_valid = df_style['path'].tolist()
```

```python
# копирование изображений для валидации в отдельную папку
val_path = 'wikiart_validation'
if not os.path.exists(val_path):
  os.makedirs(val_path)
for path in paths_valid:
  path_out = os.path.join('wikiart', path)
  shutil.copy(path_out, val_path)
```
