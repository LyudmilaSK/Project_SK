```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import time
import pandas as pd
```

```python
def write_url(url, obj):
    driver = webdriver.Chrome()
    driver.get(url)
    href = []

    
    # Проверка, является страница каталогом или ссылкой на объявление
    if 'https://auto.ru/moskva/cars/' in url:
        # если страница каталог
        while True:
            # проверка на кликабельность кнопки следующей страницы
            try:
                button_nextpage = WebDriverWait(driver, 2).until(EC.element_to_be_clickable((By.XPATH, '//a[@class="Button Button_color_white Button_size_s Button_type_link Button_width_default ListingPagination__next"]')))
            except:
                button_nextpage = None
                
            # Находим и записываем все ссылки на текущей странице
            if obj == 'model':
                items = driver.find_elements_by_xpath('//a[@class="Link ListingItemTitle__link ListingItemGroup__link"]')
            if obj == 'car':
                items = driver.find_elements_by_xpath('//a[@class="Link ListingItemTitle__link"]')
                
            for i in range(len(items)):
                href.append(items[i].get_attribute('href'))
                
            if(button_nextpage != None):
                # уходим на следующую страницу
                button_nextpage = driver.find_element_by_xpath('//a[@class="Button Button_color_white Button_size_s Button_type_link Button_width_default ListingPagination__next"]')
                href_nextpage = button_nextpage.get_attribute('href')
                driver.get(href_nextpage)
            else:
                # страниц больше нет, прерываем цикл
                break
    else:
        # если страница  - объявление
        href.append(url)
  
    # Закрываем процесс браузера:
    driver.quit()
        
    #df = pd.DataFrame(href, columns=['url']) 
        
    return href
```

```python
df_model = pd.DataFrame(write_url('https://auto.ru/moskva/cars/used/?output_type=models_list', 'model'), columns=['url']) 
```

```python
df_car = pd.DataFrame()
count_model = df_model.shape[0] 
for i in range(count_model):
    df_car = df_car.append(write_url(df_model['url'][i], 'car'), ignore_index=True)
    print('выполнен шаг' + str(i))
```
