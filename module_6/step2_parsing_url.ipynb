```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import re
```

```python
import time
```

```python
def parsing(url):


    driver = webdriver.Chrome()
    try:
        driver.get(url)
    
  
        try:
            # Нажатие кнопки "Все опции":
            WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, "//div[@class='ComplectationGroupsDesktop__cut']/span[@class='Link SpoilerLink SpoilerLink_type_default']"))).click()
        except:
            print('Кнопка "Все опции" отсутствует')

        try:
            # Запись комплектации:
            сomplectation = driver.find_elements_by_xpath("//li[@class='ComplectationGroupsDesktop__item']//descendant::*")
        
            complectation_dict = []
            for each in сomplectation[1::2]:
                complectation_dict.append(each.get_attribute("textContent"))
            complectation_dict =','.join(complectation_dict)
        except:
            complectation_dict = None


        try:
            # Запись комментария:
            comment = driver.find_elements_by_xpath("//div[@class='CardDescriptionHTML']//descendant::*")
        
            description = []
            for each in comment:
                description.append(each.get_attribute("textContent"))
            description = ','.join(description)
        except:
            description = None
        
        # Получаем содержимое html-страницы:
        page = driver.execute_script("return document.body.innerHTML;")    
    
        # Закрываем процесс браузера:
        driver.quit()
    
        soup = BeautifulSoup(page, 'html.parser')

        try:
            body_type = soup.find('li', class_='CardInfoRow CardInfoRow_bodytype').contents[-1].text.split(' ')[0]
        except:
            body_type = None

        try:
            brand = soup.find('h1', class_='CardHead__title').text.split(' ')[0]
        except:
            brand = None
    
        try:
            color = soup.find('li', class_='CardInfoRow CardInfoRow_color').contents[-1].text
        except:
            color = None

        try:
            engine_displacement = soup.find('li', class_='CardInfoRow CardInfoRow_engine').contents[-1].text.split(' ')[0]
        except:
            engine_displacement = None

        try:
            engine_power = soup.find('li', class_='CardInfoRow CardInfoRow_engine').contents[-1].text.split(' ')[3].replace('\xa0л.с.', '')
        except:
            engine_power = None

        try:
            fuel_type = soup.find('li', class_='CardInfoRow CardInfoRow_engine').contents[-1].text.split(' ')[-1]
        except:
            fuel_type = None

        try:
            image = soup.find('img', class_='ImageGalleryDesktop__image ImageGalleryDesktop__image_hidden').get('src')
        except:
            image = None
    
        try:
            mileage = soup.find('li', class_='CardInfoRow CardInfoRow_kmAge').contents[-1].text.replace('\xa0', '').replace('км', '')
        except:
            mileage = None

        try:
            model_name = soup.find('h1', class_='CardHead__title').text.split(' ')[1]
        except:
            model_name = None

        try:
            price_currency = soup.find('span', class_='OfferPriceCaption__price').get_text()[-1]
        except:
            price_currency = None

        try:
            production_date = soup.find('li', class_='CardInfoRow CardInfoRow_year').contents[-1].text
        except:
            production_date = None

        try:
            sell_id = soup.find('div', class_='CardHead__infoItem CardHead__id').get_text().replace('№ ', '')
        except:
            sell_id = None

        try:
            vehicle_transmission = soup.find('li', class_='CardInfoRow CardInfoRow_transmission').contents[-1].text
        except:
            vehicle_transmission = None

        try:
            ownerscount = soup.find('li', class_='CardInfoRow CardInfoRow_ownersCount').contents[-1].text
        except:
            ownerscount = None

        try:
            owning_time = soup.find('li', class_='CardInfoRow CardInfoRow_owningTime').get_text().replace('Владение', '')
        except:
            owning_time = None

        try:
            pts = soup.find('li', class_='CardInfoRow CardInfoRow_pts').contents[-1].text
        except:
            pts = None

        try:
            customs = soup.find('li', class_='CardInfoRow CardInfoRow_customs').contents[-1].text
        except:
            customs = None

        try:
            drive = soup.find('li', class_='CardInfoRow CardInfoRow_drive').contents[-1].text
        except:
            drive = None

        try:
            wheel = soup.find('li', class_='CardInfoRow CardInfoRow_wheel').contents[-1].text
        except:
            wheel = None

        try:
            state = soup.find('li', class_='CardInfoRow CardInfoRow_state').contents[-1].text
        except:
            state = None

        try:
            place = soup.find('span', class_='MetroListPlace__regionName MetroListPlace_nbsp').get_text()
        except:
            place = None

        try:
            price = soup.find('span', class_='OfferPriceCaption__price').get_text().replace('\xa0', '')[0:-1]
        except:
            price = None

        # Открытие html-страницы с характеристиками:
        url_characteristic = soup.find('a', class_='Link SpoilerLink CardCatalogLink SpoilerLink_type_default').get('href')
        driver_characteristic = webdriver.Chrome()
        driver_characteristic.get(url_characteristic)

        # Получаем содержимое html-страницы:
        page_characteristic = driver_characteristic.execute_script("return document.body.innerHTML;")

        # Закрываем процесс браузера:
        driver_characteristic.quit()

        soup_characteristic = BeautifulSoup(page_characteristic, 'html.parser')

        try:
            model_date = soup_characteristic.find_all('a', class_='breadcrumbs__link')[4].get_text()
        except:
            model_date = None

        try:
            number_doors = soup_characteristic.find('dt', string=re.compile('Количество дверей')).find_next_sibling('dd').get_text()
        except:
            number_doors = None

        try:
            vendor = soup_characteristic.find('dt', string=re.compile('Страна марки')).find_next_sibling('dd').get_text()
        except:
            vendor = None

        try:
            overclocking = soup_characteristic.find('dt', string=re.compile('Разгон до 100 км/ч, с')).find_next_sibling('dd').get_text()
        except:
            overclocking = None
    
        try:
            consumption = soup_characteristic.find('dt', string=re.compile('Расход топлива, л')).find_next_sibling('dd').get_text().split('/')[0]
        except:
            consumption = None

        try:
            class_auto = soup_characteristic.find('dt', string=re.compile('Класс автомобиля')).find_next_sibling('dd').get_text()
        except:
            class_auto = None

        try:
            seats = soup_characteristic.find('dt', string=re.compile('Количество мест')).find_next_sibling('dd').get_text()
        except:
            seats = None

        try:
            max_speed = soup_characteristic.find('dt', string=re.compile('Максимальная скорость, км/ч')).find_next_sibling('dd').get_text()
        except:
            max_speed = None

        parsing_unixtime = time.time()
        
        return pd.Series([url, body_type, brand, color, complectation_dict, description,
                          engine_displacement, engine_power, fuel_type, image, mileage,
                          model_name, price_currency, production_date, sell_id,
                          vehicle_transmission, ownerscount, owning_time, pts, customs,
                          drive, wheel, state, place, price, model_date, number_doors,
                          vendor, overclocking, consumption, class_auto, seats,
                          max_speed, parsing_unixtime])
    except:
        print('страница не захвачена')
        return None
```

```python
data = pd.read_csv('car.csv')
```

```python
data
```

```python
car = pd.DataFrame()
```

```python
for row in range(data.shape[0]):
    print(f'работа со строкой {row}')
    pars = parsing(data['url'][row])
    if pars is not None:
        for column in range(len(pars)):
            car.loc[row, column] = pars[column]
```

```python
car.columns = ['url', 'body_type', 'brand', 'color', 'complectation_dict', 'description',
            'engine_displacement', 'engine_power', 'fuel_type', 'image', 'mileage',
            'model_name', 'price_currency', 'production_date', 'sell_id',
            'vehicle_transmission', 'ownerscount', 'owning_time', 'pts', 'customs',
            'drive', 'wheel', 'state', 'place', 'price', 'model_date', 'number_doors',
            'vendor', 'overclocking', 'consumption', 'class_auto', 'seats',
            'max_speed', 'parsing_unixtime']
```

```python
car.to_csv('car.csv', index=False)
```
